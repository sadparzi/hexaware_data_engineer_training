```python
data = [
    (101,"Rahul",75000),
    (102,"Priya",85000),
    (103,"Amit",65000)
]

df = spark.createDataFrame(
    data,
    ["emp_id","name","salary"]
)
```

```python
df.write.format("delta") \
.save("/tmp/employees_delta")
```

```python

delta_df = spark.read.format("delta") \
.load("/tmp/employees_delta")

display(delta_df)
```

```python
df.write.format("delta") \
.saveAsTable("employees")
```
