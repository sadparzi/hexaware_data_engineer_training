```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("working with files").getOrCreate()
```

```python
df = spark.read.csv(
    r"C:\new_env\hexaware-sql\june_10\passengers.csv",
    header=True,
    inferSchema=True
)
```

```python
bookings_df = spark.read.csv(
    r"C:\new_env\hexaware-sql\june_10\booking.csv",
    header=True,
    inferSchema=True
)
bookings_df.show()
bookings_df.printSchema()
```

```python
from pyspark.sql.functions import col
bookings_df.filter(col("city").isNull()).show()

bookings_df.filter(col("provider").isNull()).show()
```

```python
drop_any_null_df = bookings_df.na.drop(how="any")
drop_any_null_df.show()
```

```python
clean_df = bookings_df.na.drop(
    subset = [
        "customer_name",
        "service_type",
        "booking_amount"
    ]
)
clean_df.show()
```

```python
filled_df = bookings_df.na.fill(
    {
        "city": "Unknown",
    }
)
filled_df.show()
```

```python
filled_df = bookings_df.na.fill(
    {
        "city": "Unknown",
        "provider": "Not Available",
        "booking_status": "Unknown",
        "payment_mode": "Not Provided",
        "booking_amount": 0
    }
)
filled_df.show()
```
