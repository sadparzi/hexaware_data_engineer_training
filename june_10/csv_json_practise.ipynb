```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, sum, avg, max, min, count, array_contains

spark = SparkSession.builder.appName("working with files").getOrCreate()
```

```python
flights_df = spark.read.csv(
    r"C:\new_env\hexaware-sql\june_10\flights.csv",
    header=True,
    inferSchema=True
)
flights_df.show()
flights_df.printSchema()
```

```python
# 1
flights_df.show(truncate=False)
```

```python
# 2
flights_df.select('passenger_name','from_city','to_city').show(truncate=False)
```

```python
# 3
flights_df.filter(col('status') == 'Confirmed').show(truncate=False)
```

```python
# 4
flights_df.filter(col('status') == 'Cancelled').show(truncate=False)
```

```python
# 5
flights_df.filter(col('status') == 'Pending').show(truncate=False)
```

```python
# 6
flights_df.filter(col('ticket_price') > 20000).show(truncate=False)
```

```python
# 7
flights_df.filter(col('to_city').isin('Dubai','Singapore','London')).show(truncate=False)
```

```python
# 8
print(flights_df.count())
```

```python
# 9
flights_df.groupBy('airline').count().show()
```

```python
# 10
flights_df.groupBy('status').count().show()
```

```python
# 11
flights_df.filter(col('status') == 'Confirmed').select(sum('ticket_price').alias('total_revenue')).show()
```

```python
# 12
flights_df.groupBy('airline').avg('ticket_price').show()
```

```python
# 13
flights_df.select(max('ticket_price')).show()
```

```python
# 14
flights_df.select(min('ticket_price')).show()
```

```python
# 15
flights_df.orderBy(col('ticket_price').desc()).show(truncate=False)
```

```python
# 16
flights_df.withColumn('tax', col('ticket_price') * 0.05).show()
```

```python
# 17
flights_df.withColumn('tax', col('ticket_price') * 0.05) \
           .withColumn('final_price', col('ticket_price') + col('ticket_price') * 0.05).show()
```

```python
# 18
flights_df.withColumn('price_category',
    when(col('ticket_price') >= 30000, 'Premium')
    .when(col('ticket_price') >= 10000, 'Standard')
    .otherwise('Budget')).show(truncate=False)
```

```python
# 19
flights_df.withColumn('price_category',
    when(col('ticket_price') >= 30000, 'Premium')
    .when(col('ticket_price') >= 10000, 'Standard')
    .otherwise('Budget')) \
    .groupBy('price_category').count().show()
```

```python
flights_df.filter(col('status') == 'Confirmed') \
          .write.mode('overwrite').parquet(r'C:\new_env\hexaware-sql\june_10\confirmed_flights.parquet')
print("Saved as confirmed_flights.parquet")
```
