```python
data = [
    (101, "Rahul Sharma", "Data Engineering", 75000, "Hyderabad"),
    (102, "Priya Reddy", "AI Engineering", 85000, "Bangalore"),
    (103, "Amit Kumar", "Data Engineering", 65000, "Mumbai"),
    (104, "Sneha Patel", "Data Science", 95000, "Chennai"),
    (105, "Farhan Ali", "Cloud Engineering", 80000, "Delhi")
]

col = [
    "employee_id",
    "name",
    "department",
    "salary",
    "city"
]
```

```python
!pip install pyspark
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Employee Data") \
    .getOrCreate()
```

```python
df = spark.createDataFrame(data, col)
```

```python
df.select("employee_id", "name").show()
```

```python
df.select("name", "salary", "city").show()
```

```python
df.drop("city").show()
```

```python
from pyspark.sql.functions import col
```

```python
df.filter(col("salary") < 80000).show()
```

```python
df.filter(col("city") == "Bangalore").show()
```

```python
df.filter(col("department") == "Data Engineering").show()
```

```python
df.filter(col("department") == "AI Engineering").show()
```

```python
df.filter(col("salary").between(70000, 90000)).show()
```

```python
df.filter(
    (col("city") == "Hyderabad") &
    (col("salary") > 70000)
).show()
```

```python
df.filter(
    (col("city") == "Bangalore") |
    (col("city") == "Chennai")
).show()
```

```python
df.filter(
    col("city").isin(
        "Hyderabad",
        "Mumbai",
        "Delhi"
    )
).show()
```

```python
df.filter(
    col("department").isin(
        "Data Engineering",
        "AI Engineering"
    )
).show()
```

```python
df.orderBy("salary").show()
```

```python
df.orderBy(col("salary").desc()).show()
```

```python
df.orderBy("name").show()
```

```python
df.orderBy("city", "salary").show()
```

```python
df = df.withColumn(
    "bonus",
    col("salary") * 0.10
)

df.show()
```

```python
df = df.withColumn(
    "tax",
    col("salary") * 0.05
)

df.show()
```

```python
df = df.withColumn(
    "annual_salary",
    col("salary") * 12
)

df.show()
```

```python
df = df.withColumn(
    "salary_after_bonus",
    col("salary") + col("bonus")
)

df.show()
```

```python
df = df.withColumnRenamed(
    "salary",
    "monthly_salary"
)

df.show()
```

```python
df = df.withColumnRenamed(
    "department",
    "team"
)

df.show()
```

```python
df = df.withColumn(
    "salary_category",
    when(col("monthly_salary") >= 85000, "High")
    .when(col("monthly_salary") >= 75000, "Medium")
    .otherwise("Low")
)

df.show()
```

```python
df.groupBy(
    "salary_category"
).count().show()
```
