# Exercise 2: product_loader

```python
from pyspark.sql.functions import avg, round as spark_round

product_data = [
    (1, 'Laptop', 'Electronics', 55000),
    (2, 'Phone', 'Electronics', 30000),
    (3, 'Desk', 'Furniture', 12000),
    (4, 'Chair', 'Furniture', 8000),
    (5, 'Headphones', 'Electronics', 5000)
]
columns = ['product_id', 'product_name', 'category', 'price']
df_products = spark.createDataFrame(product_data, columns)
display(df_products)
df_products.createOrReplaceTempView('products_view')
result = spark.sql('SELECT category, ROUND(AVG(price), 2) AS avg_price FROM products_view GROUP BY category')
display(result)
```

# Exercise 3: customer_loader

```python
customer_data = [
    (1, 'Rahul Sharma', 'Hyderabad', 'Gold'),
    (2, 'Priya Reddy', 'Bangalore', 'Silver'),
    (3, 'Amit Kumar', 'Hyderabad', 'Gold'),
    (4, 'Sneha Patel', 'Mumbai', 'Bronze'),
    (5, 'Neha Singh', 'Bangalore', 'Gold'),
    (6, 'Arjun Verma', 'Mumbai', 'Silver')
]
columns = ['customer_id', 'customer_name', 'city', 'tier']
df_customers = spark.createDataFrame(customer_data, columns)
display(df_customers)
df_customers.write.format('delta').mode('overwrite').saveAsTable('customer_table')
print('customer_table saved successfully')
```

# Exercise 3: customer_summary

```python
from pyspark.sql.functions import count

df_customers = spark.read.format('delta').table('customer_table')
summary = df_customers.groupBy('city').agg(count('customer_id').alias('customer_count')).orderBy('customer_count', ascending=False)
display(summary)
```

# Exercise 4: bank_report

```python
from pyspark.sql.functions import sum as spark_sum

account_data = [
    (1001, 'Rahul Sharma', 'Hyderabad', 'Savings', 150000),
    (1002, 'Priya Reddy', 'Bangalore', 'Current', 320000),
    (1003, 'Amit Kumar', 'Hyderabad', 'Savings', 98000),
    (1004, 'Sneha Patel', 'Mumbai', 'Savings', 210000),
    (1005, 'Neha Singh', 'Bangalore', 'Current', 450000),
    (1006, 'Arjun Verma', 'Mumbai', 'Savings', 175000)
]
columns = ['account_id', 'account_holder', 'city', 'account_type', 'balance']
df_bank = spark.createDataFrame(account_data, columns)
report = df_bank.groupBy('city').agg(spark_sum('balance').alias('total_balance')).orderBy('total_balance', ascending=False)
display(report)
```
