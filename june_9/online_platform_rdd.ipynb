```python
# 80
total = rdd.count()
freq = rdd.map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b).sortBy(lambda x: x[1], ascending=False).collect()
login_pct = round(rdd.filter(lambda x: x == 'LOGIN').count() / total * 100, 2)
vp_pct = round(rdd.filter(lambda x: x == 'VIDEO_PLAY').count() / total * 100, 2)
total_sales = sales_rdd.reduce(lambda a, b: a + b)
avg_sales = total_sales / sales_rdd.count()

print("============================================")
print("   ONLINE LEARNING PLATFORM USAGE SUMMARY")
print("============================================")
print(f"Total Activities     : {total}")
print(f"Unique Activity Types: {rdd.distinct().count()}")
print(f"LOGIN  %             : {login_pct}%")
print(f"VIDEO_PLAY %         : {vp_pct}%")
print(f"Total Sales          : {total_sales}")
print(f"Average Sales        : {round(avg_sales, 2)}")
print("--- Frequency Breakdown ---")
for act, count in freq:
    print(f"  {act}: {count}")
print("============================================")
```
