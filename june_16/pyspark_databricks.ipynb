```python
# 92
etl_df = (
    df
    .withColumn('total_amount', col('quantity') * col('price'))
    .withColumn('tax', col('quantity') * col('price') * 0.05)
    .withColumn('final_amount', col('quantity') * col('price') * 1.05)
    .withColumn('discount', col('quantity') * col('price') * 0.10)
    .withColumn('net_amount', col('quantity') * col('price') * 0.90)
)
etl_df.show(truncate=False)
```

```python
# 93
completed_df = etl_df.filter(col('status') == 'Completed')
print("Completed orders:", completed_df.count())
completed_df.show()
```

```python
# 94
revenue_df = completed_df.withColumn('revenue', col('quantity') * col('price'))
revenue_df.agg(sum('revenue').alias('total_revenue')).show()
```

```python
# 95
city_report = (
    completed_df
    .withColumn('revenue', col('quantity') * col('price'))
    .groupBy('city')
    .agg(
        count('order_id').alias('total_orders'),
        sum('revenue').alias('total_revenue'),
        avg('revenue').alias('avg_revenue')
    )
    .orderBy(col('total_revenue').desc())
)
city_report.show()
```

```python
# 96
category_report = (
    completed_df
    .withColumn('revenue', col('quantity') * col('price'))
    .groupBy('category')
    .agg(
        count('order_id').alias('total_orders'),
        sum('revenue').alias('total_revenue'),
        avg('revenue').alias('avg_revenue')
    )
    .orderBy(col('total_revenue').desc())
)
category_report.show()
```

```python
# 97
etl_df.write.mode("overwrite").parquet("/dbfs/tmp/sales_output")
city_report.write.mode("overwrite").parquet("/dbfs/tmp/city_revenue")
category_report.write.mode("overwrite").parquet("/dbfs/tmp/category_revenue")
print("Saved as Parquet")
```

```python
# 100
city_rev = (
    completed_df
    .withColumn('revenue', col('quantity') * col('price'))
    .groupBy('city')
    .agg(sum('revenue').alias('total_revenue'))
    .orderBy(col('total_revenue').desc())
)
cat_rev = (
    completed_df
    .withColumn('revenue', col('quantity') * col('price'))
    .groupBy('category')
    .agg(sum('revenue').alias('total_revenue'))
    .orderBy(col('total_revenue').desc())
)
total_rev = completed_df.withColumn('revenue', col('quantity') * col('price')).agg(sum('revenue')).collect()[0][0]
top_city = city_rev.first()[0]
top_cat = cat_rev.first()[0]
total_orders = completed_df.count()

print("=" * 50)
print("       FINAL BUSINESS REPORT")
print("=" * 50)
print(f"Total Completed Orders : {total_orders}")
print(f"Total Revenue          : {total_rev}")
print(f"Top City               : {top_city}")
print(f"Top Category           : {top_cat}")
print("--- City Revenue ---")
city_rev.show()
print("--- Category Revenue ---")
cat_rev.show()
print("=" * 50)
```
